```python
import pandas as pd
```

```python
df = pd.read_csv("ACCELERATION.csv")
```

```python
df
```

```python
dfplot = pd.DataFrame({'acceleration': list(df['acceleration']),
                  'noisyacceleration': list(df['noisyacceleration'])},
                  index=list(df['timestamp']))
dfplot.plot(title = "Acceleration vs Noisy Acceleration")
```

```python
df['speed'] = df['acceleration'].cumsum() * 0.1
df['noisyspeed'] = df['noisyacceleration'].cumsum() * 0.1
df
```

```python
dfplot2 = pd.DataFrame({'speed': list(df['speed']),
                  'noisyspeed': list(df['noisyspeed'])},
                  index=list(df['timestamp']))
dfplot2.plot(title = "Speed vs Noisy Speed")
```

```python
df['distance'] = df['speed'].cumsum() * 0.1
df['noisydistance'] = df['noisyspeed'].cumsum() * 0.1
df
```

```python
dfplot3 = pd.DataFrame({'distance': list(df['distance']),
                  'noisydistance': list(df['noisydistance'])},
                  index=list(df['timestamp']))
dfplot3.plot(title = "Distance vs Noisy Distance")
```
